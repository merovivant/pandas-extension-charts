```python
# Gera dados aleatrios de mercado
np.random.seed(42)
dates = pd.date_range(start='2024-01-01', periods=30)
base_price = 100
prices = []

# Simulao simples de Random Walk
current = base_price
for _ in range(30):
    change = np.random.uniform(-2, 2) # Variao diria
    current += change
    
    # Gera OHLC baseados no preo atual
    high = current + np.random.uniform(0, 1.5)
    low = current - np.random.uniform(0, 1.5)
    open_p = current + np.random.uniform(-0.5, 0.5)
    close_p = current # Simplificao
    
    prices.append([open_p, high, low, close_p])

df_candle = pd.DataFrame(prices, columns=['Open', 'High', 'Low', 'Close'])
df_candle['Data'] = dates

df_candle.charts.candlestick(
    date_col='Data', 
    open_col='Open', 
    high_col='High', 
    low_col='Low', 
    close_col='Close'
)
```
